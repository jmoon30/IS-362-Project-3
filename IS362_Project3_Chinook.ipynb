This notebook connects to the **Chinook** sample database (SQLite) and produces a pandas DataFrame showing **Customer LastName, FirstName, purchased Track Name, and Album Title**. Results are sorted by LastName then FirstName.

**Tables joined (5):** `Customer → Invoice → InvoiceLine → Track → Album`.

> ERD / schema: see the assignment PDF and Chinook schema page.


```python

import sqlite3
import pandas as pd

db_path = "Chinook_Sqlite.sqlite" # or whatever the database path is. This creates the results in csv file.

query = """
SELECT 
  c.LastName,
  c.FirstName,
  t.Name AS TrackName,
  a.Title AS AlbumTitle
FROM Customer AS c
JOIN Invoice AS i      ON i.CustomerId = c.CustomerId
JOIN InvoiceLine AS il ON il.InvoiceId = i.InvoiceId
JOIN Track AS t        ON t.TrackId = il.TrackId
JOIN Album AS a        ON a.AlbumId = t.AlbumId
ORDER BY c.LastName, c.FirstName, t.Name
"""

with sqlite3.connect(db_path) as con:
    df = pd.read_sql_query(query, con)

df.head()

# To save results
df.to_csv("chinook_project3_results.csv", index=False)
df.head(20)
```
